Getting the percentages asked for in the BRES survey

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
from functions import reorder_file_paths
```

```python
# Load in an image edited by a vegetative index
img = cv2.imread('results/cropped_rgbvi_gray/image_25.jpg', 0)

plt.hist(img.flat, bins=100)
```

```python
# Determine a threshold that will differentiate between green and not
binary = img >= 150

plt.imshow(binary, cmap='gray')
```

Calculate how many pixels are within the threshold

```python
# Get the total amount of pixels
total_pixels = img.size

green_pixels = np.count_nonzero(binary)

percent_green = green_pixels / total_pixels

print(percent_green)
```

Now use a for loop to get one for every image

```python
# Establish path
folder = 'results\\cropped_rgbvi_gray\\*'

# Get files in order
files = reorder_file_paths(folder)

print(files)

# Set up list for appending later
percent_green = []
```

```python
for file in files:
    # Load in image
    img = cv2.imread(file, 0)

    # Determine a threshold that will differentiate between green and not
    binary = img >= 150

    # Get the total amount of pixels
    total_pixels = img.size

    # Count how many pixels are green
    green_pixels = np.count_nonzero(binary)

    # Calculate percentage of green pixels
    percent_green_val = green_pixels / total_pixels

    percent_green.append(percent_green_val)
```
